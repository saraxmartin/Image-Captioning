```python
import pandas as pd


df = pd.read_csv("data/captions.csv")

df.head()
```

```python
# Remove trailing semicolons from all column values
df = df.applymap(lambda x: str(x).rstrip(';') if isinstance(x, str) else x)
df.head(20)
```

```python
df['image_name'] = df['image_name'].str.strip()
df.head()
```

```python
import nltk
import os
# Path to your virtual environment's nltk_data directory
nltk_data_path = os.path.join('venv', 'nltk_data')

# Ensure the directory exists
os.makedirs(nltk_data_path, exist_ok=True)

# Add this path to NLTK's data paths
nltk.data.path.append(nltk_data_path)

# Download punkt
nltk.download('punkt', download_dir=nltk_data_path)
```

```python
def preprocess_caption(caption):
    # Lowercase
    caption = caption.lower()
    # Tokenize
    tokens = nltk.word_tokenize(caption)
    # Join tokens
    return " ".join(tokens)

# Apply preprocessing
df['cleaned_caption'] = preprocess_caption(df[' comment;'])
#df['cleaned_caption'] = df[' comment;'].apply(preprocess_caption)
df['cleaned_caption'].head()
```
