```python
import pandas as pd
import os
df = pd.read_csv("data/captions.csv")

df.head()
```

```python
def other_csv(text,csv, csv1, headers):  # Default 'a', can be modified
    l = []
    l2 = []
    for row in text:
        if chr(34) in row[0]:
            l.append(row)
        else:
            if chr(34) in row:
                l.append(row)
            else:
                l2.append(row)

    
    df = pd.DataFrame(l, columns=headers)
    df.to_csv(csv, index=False)
    print(f"CSV created: {csv}")
    df_cleaned = pd.DataFrame(l2, columns=headers)
    df_cleaned.to_csv(csv1, index=False)
    return df_cleaned

directory = 'data'
file_path = os.path.join(directory, 'other_captions.csv')
file_path1 = os.path.join(directory, 'captions_cleaned.csv')

df_cleaned = other_csv(df.values.tolist(), file_path, file_path1, df.columns)
print(len(df_cleaned))
df_cleaned.head()
```

```python
nan_count = df_cleaned.isna()
print(f"Number of NaN values in 'comment;' column: {nan_count}")
```
